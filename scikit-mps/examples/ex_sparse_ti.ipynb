# MPSlib: Using a sparse training image

An example of using a sparse training image with MPSlib.
Missing values are indicated by `np.nan` (or the sentinel value `-997799`),
and are ignored during the training image scan.

```python
import numpy as np
import matplotlib.pyplot as plt
import mpslib as mps
```

## Setup

```python
# Initialize MPSlib using the mps_snesim_tree algorithm and a simulation grid of size [128, 128, 1]
O = mps.mpslib(method='mps_snesim_tree', simulation_grid_size=[128, 128, 1])
#O = mps.mpslib(method='mps_genesim', simulation_grid_size=[128, 128, 1])

O.par['debug_level'] = -1
O.par['n_cond'] = 25
O.par['n_real'] = 32
O.par['n_threads'] = 8
```

## Setup sparse training image

All values of `np.nan` (or `-997799`) will be treated as missing data.

```python
#TI, TI_filename = mps.trainingimages.strebelle(di=2, coarse3d=1)
TI, TI_filename = mps.trainingimages.strebelle(di=2, coarse3d=0)
nanval = np.nan
TI2 = TI.copy()
TI2[30:60, 10:40] = nanval

n = 10000
ixx = np.random.randint(0, TI.shape[0], n)
iyy = np.random.randint(0, TI.shape[1], n)
for i in range(n):
    TI2[iyy[i], ixx[i]] = nanval

O.ti = TI2
O.par['ti_fnam'] = 'ti_sparse.ti'

plt.pcolor(O.ti[:, :, 0].T, cmap='jet')
plt.colorbar()
plt.title('Sparse training image')
```

## Set hard data from the non-missing TI values

```python
d_hard = np.empty((0, 4), int)
for ix in np.arange(O.ti.shape[0]):
    for iy in np.arange(O.ti.shape[1]):
        if O.ti[iy, ix, 0] > -1:
            d_hard = np.append(d_hard, np.array([[iy, ix, 0, O.ti[iy, ix, 0]]]), axis=0)
O.d_hard = d_hard
print(O.d_hard.shape)
print(O.ti.shape)
print(125 * 125)
```

## Run simulation

```python
#O.run()
O.run_parallel()
```
